# 1. Сгруппировать данные по кампаниям и оценить их эффективность.


К сожалению, склеить два датафрейма методом merge или join по общему ключу не удалось по технической причине (есть скрины).
Для таких больших таблиц не хватает оперативной памяти компьютера. 
Поэтому приведу анализ по отдельности: сначала e-mail, затем viber.(Относится ко всем пунктам задания)

##### Будем считать эффективностью - максимальную конверсию: клика e-mail к отправке e-mail и клик на основную ссылку в Viber к отправке сообщений в Viber (поскольку в задании не дан критерий эффективности)

# E-mail:

```python
#Вычислили конверсию в клик (0 - не перешел по ссылку в почте, 1 - перешел по ссылке в почте)
df_email_optimal['Conversion'] = df_email_optimal['EMAIL_CLICKED'] / df_email_optimal['EMAIL_SENT']
```

```python
df_email_optimal['Conversion'].sum()
```

```python
#Функция подсчета конверсии в процентах для применения через apply после группировки
def CR(row):
    df = {
        'CR E-mail Clicked': (row['Conversion'].sum() / row['EMAIL_SENT'].sum())*100, 
         'Scale of marketing campaign': (row['EMAIL_SENT'].count())       
    }
    return pd.Series(df) 
```

```python
def CR_only(row):
    df = {
        'CR E-mail Clicked': (row['Conversion'].sum() / row['EMAIL_SENT'].sum())*100,        
    }
    return pd.Series(df) 
```

```python
#Не до конца понял, рекламная компания == сегмент или ID? поэтому учтем и то, и то
df_email_campaign_id = df_email_optimal.groupby(['CAMPAIGN_SEGMENT', 'CAMPAIGN_ID']).apply(CR)
df_email_campaign_id
```

```python
#Средние значения
df_email_campaign_id.groupby('CAMPAIGN_SEGMENT').agg({'CR E-mail Clicked': 'mean'})
```

```python
df_email_segment = df_email_optimal.groupby(['CAMPAIGN_SEGMENT']).apply(CR)
df_email_campaign_id_CR_ONLY = df_email_optimal.groupby(['CAMPAIGN_ID']).apply(CR_only)
df_email_segment_CR_ONLY = df_email_optimal.groupby(['CAMPAIGN_SEGMENT']).apply(CR_only)
```

### Наибольшая масштабность у кампании с идентификатором - 6511, было отправлено 2418 e-mail

```python
df_email_campaign_id.plot(kind = "barh", figsize = (12,8))
```

### Видно, что масштаб рекламной кампании в сегменте "pediatr" был самый большой (наибольшее количество отосланных email)

```python
df_email_segment.plot(kind = "barh")
```

### Наиболее эффективной является кампания с идентификатором 5910 с показателем конверсии 6.93%

```python
df_email_campaign_id_CR_ONLY.plot(kind = "barh", figsize = (12,8))
```

### Видно, что наиболее эффективный сегмент рекламной кампании по проценту открытых е-мейлов - "Allerg"  

```python
df_email_segment_CR_ONLY.plot(kind = "barh")
```

```python
df_viber_optimal['Conversion'].sum()
```

```python
def CR_only_viber(row):
    df = {
        'CR Viber read': (row['Conversion'].sum() / row['VIBER_SENT'].sum())*100       
    }
    return pd.Series(df) 
```

# 3. Посчитать показатель вовлеченности(ES) для каждого пользователя, используя весовые коэффициенты: открытие письма = 1, клик в письме = 2, прочтение вайбера = 1, клик в вайбере = 2

```python
def ES_email(row):
    df = {
        'ES': ((row['EMAIL_OPENED'] + row['EMAIL_CLICKED']*2).sum() / row['EMAIL_SENT'].sum())*100    
    }
    return pd.Series(df) 
```

```python
df_email_es = df_email_optimal.groupby('CONTACT_ID').apply(ES_email).sort_values(by = 'ES', ascending = False)
df_email_es
```

```python
def ES_viber(row):
    df = {
        'ES': ((row['VIBER_READ'] + row['MAIN_LINK_CLICKED']*2).sum() / row['VIBER_SENT'].sum()) * 100,    
    }
    return pd.Series(df) 
```

```python
df_viber_es = df_viber_optimal.groupby('CONTACT_ID').apply(ES_viber).sort_values(by = 'ES', ascending = False).dropna()
df_viber_es
```
